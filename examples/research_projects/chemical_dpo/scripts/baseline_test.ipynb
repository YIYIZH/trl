```python
from tqdm import tqdm
```

```python
data = pd.read_csv('test_prompt.csv', encoding='gb18030')
data['answer'] = ''
```

```python
for i in tqdm(range(len(data))):
  user_prompt = data.at[i,'prompt']
  system_prompt = """You are a helpful assistant"""

  completion = client.chat.completions.create(
    model = "gpt-3.5-turbo",
    messages = [
      {"role": "system", "content": system_prompt},
      {"role": "user", "content": user_prompt}
    ],
  )
  baseline_answer = completion.choices[0].message.content
  data.at[i,'answer'] = baseline_answer
  print(baseline_answer)
  print('----------------------------------------------------------------------------------------------------')
```

```python
data.to_csv('model_result_gpt.csv', index=False)
```
